```python
# %load cal_kendall_coefficient
import torch
from scipy.stats import kendalltau

# 文件路径
tf_file_path = "/home/zzy/TF_TAS/all_output/5/ranking_checkpoint-1000cands.pth"
af_file_path = "/home/zzy/AutoFormer/all_output/5/af_ranking-1000cands.pth"

# 从pth文件加载数据
tf_cand_info = torch.load(tf_file_path)
af_cand_info = torch.load(af_file_path)

tf_rankings = tf_cand_info['ranking']
af_ranking = af_cand_info['af_ranking']

# 提取排名list

af_rank_list = []
tf_ranking_lists = {}

for k in tf_rankings:
    print(k)
```

```python
af_list = []
tf_lists = {}
for res in af_cand_info['af_all_res']:
    af_list.append(res['acc1'])

#初始化,{proxy_name : null_list}
indicator = tf_cand_info['all_res'][0]['indicator']
for key in indicator:
    n_list = []
    tf_lists[key] = n_list

for res in tf_cand_info['all_res']:
    indicator = {}
    indicator = res['indicator']
    for key in indicator:
        tf_lists[key].append(indicator[key])

print('af_list')
print(af_list)
for key in tf_lists:
    print(key)
    print(tf_lists[key])
```

```python
#使用原始数据计算tau

kendall = {}

print('候选数量:{}'.format(len(af_list)))
for k in tf_lists:
    kendall[k], _ = kendalltau(tf_lists[k],af_list)

for k in kendall:
    print('key:{},value:{}'.format(k,kendall[k]))
```
